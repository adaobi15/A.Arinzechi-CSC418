```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Predefined user database (Example UMC members)
umc_members = {
    "enase": {"surname": "Okonedo", "image": "image3.jpg"},
    "chris": {"surname": "Ogbechie", "image": "image4.jpg"},
    "ikechukwu": {"surname": "Obiaya", "image": "image5.jpg"},
    "sola": {"surname": "Oni", "image": "image6.jpg"},
    "darlington": {"surname": "Agbholor", "image": "image7.jpg"},
}

def calculate_password(surname):
    """Calculate password as the sum of the number of letters in the surname"""
    return len(surname.strip())  # Remove any leading/trailing spaces

def authenticate_user(username, password):
    """Check if the username and password match the records"""
    username = username.lower().strip()  # Convert input username to lowercase and remove spaces
    
    if username in umc_members:
        correct_password = calculate_password(umc_members[username]["surname"])
        if password == correct_password:
            return umc_members[username]["image"], username.capitalize()
        else:
            print("‚ùå Incorrect password.")
            return None, None
    else:
        print("‚ùå Username not found.")
        return None, None

def display_profile(image_path, name):
    """Display the user's profile image with their name using Matplotlib"""
    if os.path.exists(image_path):
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for Matplotlib
        
        plt.figure(figsize=(5, 5))  # Set figure size
        plt.imshow(image)
        plt.title(f"Profile: {name}", fontsize=14)  # Display user name as title
        plt.axis("off")  # Hide axes
        plt.show()
    else:
        print("‚ö†Ô∏è Profile image not found!")

# User Login Process
print("üîê UMC Facial Recognition Login System")

username = input("Enter your first name: ").strip()
surname = input("Enter your surname: ").strip()
password = calculate_password(surname)

print(f"üîç Authenticating {username.capitalize()}...")

image_path, user_name = authenticate_user(username, password)

if image_path:
    print(f"‚úÖ Login successful! Displaying {user_name}'s profile image...")
    display_profile(image_path, user_name)
else:
    print("‚õî Access Denied.")
```
