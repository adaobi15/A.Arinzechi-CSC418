```python
import cv2

image1 = cv2.imread('image1.jpg')
image2 = cv2.imread('image2.jpg')

# Ensure images are loaded
if image1 is None or image2 is None:
    print(" Error: One or both images could not be loaded. Check the file paths.")
else:
    # Resize images to the same size
    image1 = cv2.resize(image1, (500, 400))
    image2 = cv2.resize(image2, (500, 400))

    # Perform weighted addition
    addImage = cv2.addWeighted(image1, 0.5, image2, 0.6, 0)

    # Show result
    cv2.imshow('Weighted Image', addImage)

    # Wait for key press (27 = ESC key), then close
    key = cv2.waitKey(0) & 0xFF
    if key == 27:  # Check if ESC is pressed
        cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np 

# Load the image
image = cv2.imread('image1.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB for Matplotlib

# Display original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)

# Adjust brightness & contrast
brightness = 5
contrast = 1.5
image2 = cv2.addWeighted(image, contrast, np.zeros(image.shape, dtype=image.dtype), 0, brightness)

# Save modified image
cv2.imwrite('image1_modified.jpg', cv2.cvtColor(image2, cv2.COLOR_RGB2BGR))  # Convert back to BGR for saving

# Display modified image
plt.subplot(1, 2, 2)
plt.title("Brightness & Contrast")
plt.imshow(image2)

plt.show()
```

```python
def grayscale(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
def adjust_brightness(image, alpha=1.5, beta=30):
    return cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
```

```python
def sharpen(image):
    kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
    return cv2.filter2D(image, -1, kernel)
```

```python
def denoise(image):
    return cv2.GaussianBlur(image, (5, 5), 0)
```

```python
def process_image(image_path, enhancement_type):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    if enhancement_type == "1":
        enhanced_image = grayscale(image)
        cmap, title = "gray", "Grayscale"
    elif enhancement_type == "2":
        enhanced_image = adjust_brightness(image)
        cmap, title = None, "Brightness Adjusted"
    elif enhancement_type == "3":
        enhanced_image = sharpen(image)
        cmap, title = None, "Sharpened"
    elif enhancement_type == "4":
        enhanced_image = denoise(image)
        cmap, title = None, "Denoised"
    else:
        print("Invalid choice.")
        return

    # Display images
    plt.subplot(1, 2, 1)
    plt.imshow(image)
    plt.title("Original")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(enhanced_image, cmap=cmap)
    plt.title(f"Enhanced - {title}")
    plt.axis("off")

    plt.show()
```
